```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
cancer = pd.read_csv('cancer.csv')
cancer.head()
```

```python
cancer = cancer.drop(columns = ['Unnamed: 32','id'])
cancer['diagnosis'] = cancer['diagnosis'].map({'M':1 ,'B':0})
```

```python
cancer.columns
```

```python
cancer.describe()
```

```python
x = cancer.loc[:,cancer.columns !='diagnosis']
y = cancer['diagnosis']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(np.asarray(x_train))
x_test = scaler.fit_transform(np.asarray(x_test))
```

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['linear', 'poly', 'rbf']}
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
grid = GridSearchCV(SVC(), param_grid, verbose = 3)
 
grid.fit(x_train, y_train)

```

```python
pd.DataFrame(grid.cv_results_)
```

```python
best = grid.best_score_
best
```

```python
par = grid.best_params_
```
