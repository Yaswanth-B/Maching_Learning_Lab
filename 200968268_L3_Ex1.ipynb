```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv('train.csv')
titanic.head()
```

```python
titanic.count()
```

```python
titanic.isnull().sum()
```

```python
titanic.dropna(inplace=True)

titanic.isnull().sum()
```

```python
titanic.head(20)
```

```python
titanic['Family'] = titanic['Parch'] + titanic['SibSp']
titanic['has_cabin'] = np.where(titanic['Cabin'].isnull(),1,0)
titanic =titanic.drop(columns=['Ticket', 'PassengerId', 'Name', 'Parch', 'SibSp', 'Cabin'])
titanic['Sex'] = titanic['Sex'].map({'male':0, 'female':1})
titanic['Embarked'] = titanic['Embarked'].map({'C':0, 'Q':1, 'S':2})
```

```python
titanic.head()
```

```python
x_train = titanic.loc[:, titanic.columns != 'Survived']
y_train = titanic['Survived']
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('nb',GaussianNB())
])
```

```python
clf.fit(x_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(GaussianNB(),x_train, y_train, cv=5)
```

```python
score.mean()
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
y_pred = cross_val_predict(clf, x_train, y_train, cv=10)
cm = confusion_matrix(y_train, y_pred)
```

```python
tn = cm[0,0]
fp = cm[0,1]
fn = cm[1,0]
tp = cm[1,1]
acc = (tp+tn)/(tp+tn+fp+fn)
tpr = tp/(fn+tp)
tnr = tn/(tn+fp)

print("accuracy = ", acc)
print("TPR = ", tpr)
print("TNR = ", tnr)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
falsepr, truepr, _ = roc_curve(y_train, y_pred)
plt.plot(falsepr,truepr)
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.show()
```

```python
r =  roc_auc_score(y_train,y_pred)
r
```

```python
#in general, a curve moving towards the top and left of and roc curve means the classifier is better 
#area under the curve is directly proportional to how good the classifier is
#a 45 degree line corresponds to random predictions
#The area under the curve is 0.78, indicating it is a decently accurate model
```
