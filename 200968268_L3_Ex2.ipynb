```python
def cont_to_exp(x):
    if type(x) is str:
        x = x.replace('\\','')
        for key in contractions:
            value = contractions[key]
            x = x.replace(key, value)
        return x
    else:
        return x
```

```python
cl.head()
```

```python
cl["Review Text"] = cl["Review Text"].apply(lambda x:cont_to_exp(x))
cl["Review Text"]= cl["Review Text"].str.lower()
```

```python
cl.head()
```

```python
cl["Review Text"] = cl["Review Text"].str.translate(str.maketrans('', '', string.punctuation))
```

```python
cl.head()
```

```python
cl =cl.drop(columns=['Clothing ID'])
#cl =cl.drop(columns=['Clothing ID', 'Age','Rating', 'Positive Feedback Count'])
```

```python
x = cl["Review Text"]

y = cl['Recommended IND']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
y_pred = clf.predict(x_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
tn = cm[0,0]
fp = cm[0,1]
fn = cm[1,0]
tp = cm[1,1]

precision = tp/(tp+fp)
recall = tp/(tp+fn)
f1  = 2*(precision*recall/(precision+recall))

met = ['precision', 'recall', 'F1 score']
val = [precision, recall, f1]
d = {'Accuracy Metrics': met, 'Values': val}
df = pd.DataFrame(data=d)
```
