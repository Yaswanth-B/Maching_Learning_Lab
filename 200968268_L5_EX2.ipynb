```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house = pd.read_csv('HousePrice.csv')
```

```python
house.head()
```

```python
#dropping columns with null values
a = []
for col in house.columns:
    if int(house[col].isnull().sum())>0:
        a.append(col);
        house = house.drop([col], axis =1)
house = house.drop(columns = 'Id')
```

```python
#getting dummies for all categorical attributes
for col in house.columns:
    if(house[col].dtype.name=='object'):
        temp = pd.get_dummies(house[col])
        house = pd.concat([house, temp], axis=1)
        house = house.drop(columns = col)
       
    
```

```python
x = house.loc[:, house.columns!='SalePrice']
y = house['SalePrice']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)

scaled_x = scaler.transform(x_train)
```

```python
from sklearn import linear_model
lasso = linear_model.Lasso(alpha=0.1)
lasso.fit(scaled_x, y_train)
```

```python
y_pred = lasso.predict(scaled_x)
from sklearn.metrics import r2_score
score = r2_score(y_pred, y_train)
score
```

```python
from sklearn import linear_model
ridge = linear_model.Ridge(alpha=0.1)
ridge.fit(scaled_x, y_train)
```

```python
y_pred1 = ridge.predict(scaled_x)
from sklearn.metrics import r2_score
score1 = r2_score(y_pred1, y_train)
score1
```

```python
#scaling and checking accuracy on test data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_test)

scaled_x_test = scaler.transform(x_test)
y_pred_ltest = lasso.predict(scaled_x_test)
y_pred_rtest = ridge.predict(scaled_x_test)

scorel = r2_score(y_pred_ltest, y_test)
scorer = r2_score(y_pred_rtest, y_test)
print(scorel, scorer)
```

```python
data = {'Metrics':["Accuracy"],'Lasso Regression': [scorel*100], 'Ridge Regression': [scorer*100]}  
result = pd.DataFrame(data)
```
